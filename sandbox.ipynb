```python
import requests
import pandas as pd
from credentials import key
from datetime import datetime
```

```python
lat = 40.711017
long = -74.016937
```

```python
url = f'https://api.darksky.net/forecast/d131dc9bc523abbcd427cb5bed55182e/{lat},{long}'
```

```python
r = requests.get(url)
```

```python
df = pd.DataFrame(eval(r.text)['hourly']['data'])
```

```python
df['timestamp'] = [datetime.utcfromtimestamp(ts) for ts in df['time']]
df['date'] = [i.date() for i in df['timestamp']]
```
